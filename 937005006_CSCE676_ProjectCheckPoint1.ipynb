# (D) Exploratory Data Analysis (Selected Dataset Only)

Perform EDA – Ex:
- Distribution of basket sizes
- Frequency of top items
- Sparsity of item co-occurrence
- Temporal gaps between transactions
- Initial observations motivating external techniques

```python
# load in libraries I'll need
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# load in multiple result files to create dataset
youtube_search_results_path = "/Users/jsalguero/Downloads/search-results"
list_of_search_result_files = os.listdir(youtube_search_results_path) # 12 zip files

# combine result files into one dataset
search_result_dfs = []
for i in list_of_search_result_files:
    # save information from the file names
    # city-bot_num-top50-search-results
    if i.endswith(".csv"): # don't want to do this for any hidden files in the folder
        split_filename = i.split("-")
        #print(i)
        city = split_filename[0]
        bot_num = split_filename[1]
        # 3 cities from USA, 3 cities from South Africa
        if(city == "houston" or city == "losangeles" or city == "jacksonville"):
            country = "USA"
        else:
            country = "South Africa"

        # add the columns to each dataset
        curr_df = pd.read_csv(youtube_search_results_path + "/" + i)
        curr_df["city"] = city
        curr_df["bot_num"] = bot_num
        curr_df["country"] = country

        # to keep all individual dfs and combine them later, add each df into a list
        search_result_dfs.append(curr_df)

# combine into one dataset
youtube_search_result_df = pd.concat(search_result_dfs, ignore_index=True)
print(youtube_search_result_df)
```

### 1) Distribution of basket sizes

To understand the distribution of videos by search query, I grouped videos into basket sizes; each search query is a transaction and each basket is the number of unique videos for that query. The resulting histogram ranges from less than 50 to over 600 videos. This suggests that there are some queries that only show very specific videos, while other queries return a much broader range of videos. The large range of basket sizes is appropriate because the combined dataset includes multiple audit days and multiple locations. 

```python
# 1) Distribution of basket sizes
# basket sizes based on search query
# basket size = number of vides per query
basket_sizes = youtube_search_result_df.groupby("Search Query")["Video ID"].nunique().reset_index(name="basket_size")
# show basket sizes in a graph, to visualize better
plt.figure()
basket_sizes["basket_size"].hist(bins=50)
plt.xlabel("Basket Size (Number of Videos)")
plt.ylabel("Count")
plt.title("Distribution of Basket Sizes per Search Query")
plt.show()
```

### 2) Frequency of top items

Next, I wanted to visualize the frequency of the top videos and channels from the dataset. Top videos had a lot of video titles focused on controversial COVID-19 topics such as lab-leak theories, vaccines, and conspiracy-related claims. There were also a clear divide between the #1 channel WION and the rest of the top channels. This suggests that certain channels and narratives are repeatedly given by YouTube’s search algorithm across different queries and locations.

```python
# 2) Frequency of top items
# top videos & channels
top_10_videos = (youtube_search_result_df["Video Title"].value_counts(ascending=False))[:10]
#print(top_10_videos)
# visualize with barchart
top_10_videos.plot(kind="barh")
plt.xlabel("Frequency")
plt.ylabel("Video Title")
plt.title("Top 10 Most Frequent Videos")
plt.show()

# top channels
top_10_channels = youtube_search_result_df["Channel Name"].value_counts(ascending=False)[:10]
#print(top_10_channels)
# visualize with barchart
top_10_channels.plot(kind="barh")
plt.xlabel("Frequency")
plt.ylabel("Channel")
plt.title("Top 10 Most Frequent Channels")
plt.show()
```

### 3) Sparsity of item co-occurrence

Then, I analyzed the co-occurrence between search queries, search filters, and video titles. I counted how often each combination appeared, and then measured sparsity by calculating the proportion of co-occurrences that only appeared once.

The results show that most items are repeated videos rather than unique, indicating low sparsity. Many videos appear multiple times for the same query across different audit runs, locations, and sorting filters. This suggests that YouTube’s search algorithm consistently surfaces the same videos for certain queries, leading to repeated exposure and higher spreading of specific information. 

```python
# 3) Sparsity of item co-occurrence
#cocurrence matrix - similar to homework 2
cocurrence_df = youtube_search_result_df[["Search Query", "Search Filter Sort-By", "Video Title"]].copy()
cocurrence_counts = cocurrence_df.value_counts().reset_index(name="count")
cocurrence_counts = cocurrence_counts.sort_values("count", ascending = False)

# calculate sparsity
# are there a lot of repeats? or mostly unique?
occurred_once = (cocurrence_counts["count"] == 1).sum()
#print(occurred_once)
is_sparse = occurred_once/len(cocurrence_counts["count"]) # unique / total combos
#print(is_sparse)

if(is_sparse >= 0.80):
    print("Most items are unique.")
else:
    print("Most items are repeats.")
    
# What are the most popular Search Query x Video Title?
print("10 Most Popular Search Query x Search Filter Sort-By x Video Title: ", cocurrence_counts.head(10))

# What are the least popular Search Query x Video Title?
print("10 Least Popular Search Query x Search Filter Sort-By x Video Title: ", cocurrence_counts.tail(10))
```

### 4) Temporal gaps between transactions

To look at how the data changes over time, I analyzed the number of videos collected on each audit day using the run number as a representative for time. Overall, the number of videos per day stays fairly consistent, with only small fluctuations across the audit period. This suggests that the data collection process was stable and that no single day had an unusually large impact on the dataset. Since there are no major spikes or drops, the search results appear to be relatively steady over time, which makes the dataset reliable for further analysis during the semester.

```python
# 4) Temporal gaps between transactions
# based on Run #, which is the day of the data collection
# how many videos in each audit day?
videos_per_day = youtube_search_result_df.groupby("Run #")
videos_per_day_counts = videos_per_day["Video ID"].count()

# visualize videos per day
plt.figure()
videos_per_day_counts.plot(marker="o")
plt.xlabel("Audit Day")
plt.ylabel("Number of Videos")
plt.title("Number of Videos Per Day")
plt.show()
```

### 5) Compare Countries & Their Level of Misinformation

To see which country had more misinformation in search results, I looked at videos labeled as supporting misinformation and normalized the counts by the total number of videos collected for each country. This normalization is important since the U.S. and South Africa do not have the exact same number of videos in the dataset. After normalizing, the U.S. has a slightly higher proportion of misinformation-related videos than South Africa. The difference is relatively small, but it suggests that users searching from the U.S. may be exposed to slightly more misinformation during the audit period.

```python
# which country has more misinformation in search queries?
# annotation: 1 = supporting misinformation, 0 = neutral, -1 = opposing misinformation
misinformation = youtube_search_result_df[youtube_search_result_df["annotation"] == 1]

# count by country 
misinformation_country_counts = misinformation["country"].value_counts()
#print(misinformation_country_counts)

# both countries don't have same number of videos
video_counts_country = youtube_search_result_df["country"].value_counts()
#print(video_counts_country) # US has more videos in the combined dataset than South Africa

# final misinformation counts
final_misinformation = misinformation_country_counts/video_counts_country
print("Percent of Videos Leading Towards Misinformation Per Country:")
print(final_misinformation)
```
